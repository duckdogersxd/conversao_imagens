```python
# Etapa 2: Importação das bibliotecas
import cv2
import numpy as np
import urllib.request
import matplotlib.pyplot as plt

# Etapa 3: Baixa uma imagem colorida (exemplo: Lena)
url = 'https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png'
resp = urllib.request.urlopen(url)
img_arr = np.asarray(bytearray(resp.read()), dtype=np.uint8)
imagem_colorida = cv2.imdecode(img_arr, cv2.IMREAD_COLOR)

# Verifica se a imagem foi carregada
if imagem_colorida is None:
    raise Exception("Erro ao carregar a imagem!")

# Etapa 4: Conversão para escala de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)

# Etapa 5: Binarização com threshold (limiar = 127)
limiar = 127
_, imagem_binarizada = cv2.threshold(imagem_cinza, limiar, 255, cv2.THRESH_BINARY)

# Etapa 6: Exibição dos resultados com matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB))
plt.title('Imagem Colorida')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Escala de Cinza')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(imagem_binarizada, cmap='gray')
plt.title('Imagem Binarizada')
plt.axis('off')

plt.tight_layout()
plt.show()
```
